```python

# Load all matches for Premier League 2015/16
matches = sb.matches(competition_id=2, season_id=27)

# Filter matches where Leicester City is either home or away
leicester_matches = matches[
    (matches['home_team'] == 'Leicester City') |
    (matches['away_team'] == 'Leicester City')
]

# List of match IDs to load
leicester_match_ids = leicester_matches['match_id'].tolist()
print(f"Leicester played {len(leicester_match_ids)} matches")
```

```python
from kloppy import statsbomb

# Load datasets for each Leicester City match
datasets = []
for match_id in leicester_match_ids:
    try:
        ds = statsbomb.load_open_data(match_id=match_id)
        print(f"Loaded match {match_id} with {len(ds.events)} events")
        datasets.append(ds)
    except Exception as e:
        print(f"Failed to load match {match_id}: {e}")
```

```python
import pandas as pd

# Convert events to DataFrame
def events_to_df(ds):
    rows = []
    for e in ds.events:
        row = {
            "event_type": getattr(e, "event_name", type(e).__name__),
            "team_name": e.team.name if e.team else None,
            "player_name": e.player.name if e.player else None,
            "start_x": e.start.x if hasattr(e, "start") else None,
            "start_y": e.start.y if hasattr(e, "start") else None,
            "end_x": e.end.x if hasattr(e, "end") else None,
            "end_y": e.end.y if hasattr(e, "end") else None,
        }
        rows.append(row)
    return pd.DataFrame(rows)

# Combine all Leicester matches
leicester_df = pd.concat([events_to_df(ds) for ds in datasets], ignore_index=True)
print(f"Total Leicester events: {len(leicester_df)}")
```

```python
# Check unique event types
print(leicester_df["event_type"].unique())
```

```python
# Save event Data for Leicester City Games to CSV
leicester_df.to_csv("leicester_2015_16_events_readable.csv", index=False)


```
